```
import pandas as pd

df = pd.read_csv("Combined_dataset.csv")

print("First 5 Rows")
print(df.head())

print("\nLast 5 Rows")
print(df.tail())

print("\nDataset Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

df["discount"] = df["discount"].fillna(df["discount"].mean())
df["seller_name"] = df["seller_name"].fillna("Not Available")
df["videos"] = df["videos"].fillna("Not Available")

new_df = df[["product_id","title","category","rating"]]
print(new_df.head())

high_rating = df[df["rating"] > 4]
print(high_rating[["title","rating"]].head())

df = df.drop_duplicates()

df["rating_score"] = df["rating"] * 10

print(df.describe())

df.to_csv("cleaned_combined_dataset.csv", index=False)

print("Cleaned Dataset Saved Successfully")
```
